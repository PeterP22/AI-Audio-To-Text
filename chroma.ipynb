```python
from langchain.chains.question_answering import load_qa_chain
from langchain.chat_models import ChatOpenAI
from langchain.document_loaders import DirectoryLoader
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
import os
from gtts import gTTS
from IPython.display import Audio, display

os.environ["OPENAI_API_KEY"] = ""

# initializing the embeddings and models
embeddings = OpenAIEmbeddings()
llm = ChatOpenAI()
directory = "/content/files"

# Load, split documents, and initialize Chroma
documents = load_docs(directory)
docs = split_docs(documents)
db = Chroma.from_documents(documents=docs, embedding=embeddings)
chain = load_qa_chain(llm, chain_type="stuff")

def load_docs(directory):
    loader = DirectoryLoader(directory)
    return loader.load()

def split_docs(documents, chunk_size=500, chunk_overlap=50):
    splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    return splitter.split_documents(documents)

def get_answer(query):
    similar_docs = db.similarity_search(query, k=2)
    answer = chain.run(input_documents=similar_docs, question=query)
    return answer

def synthesize_and_play_answer(answer):
    tts = gTTS(text=answer, lang='en')
    tts.save("/content/answer.mp3")
    display(Audio("/content/answer.mp3", autoplay=True))

print("Private Q&A chatbot")
prompt = input("Enter your query here: ")

if prompt:
    answer = get_answer(prompt)
    print(f"Answer: {answer}")
    synthesize_and_play_answer(answer)
```
